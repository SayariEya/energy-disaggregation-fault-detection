```python
import pandas as pd
import numpy as np

# Filtrer les données pour ne garder que les 'sn' qui se terminent par '002'
df = df[df['sn'].str.endswith('002')]

# Convertir l'horodatage (timestamp) en format de date lisible
df['date'] = pd.to_datetime(df['date'], unit='ms')

# Remplacer les NaN par True
df = df.fillna(True)

# Afficher les 5 premières lignes du DataFrame ordonné
df.head()
```

```python
def flatten_data(df):
    records = []
    for _, row in df.iterrows():
        base = {
            '_id': row['_id'],
            'date': row['date'],
            'sn': row['sn'],
            'sft_ver': row['sft_ver'],
            'meter': row['meter'],
            'nbr_channel': row['nbr_channel'],
            '_class': row['_class']
        }
        for ed in row['electric_data']:
            ed_base = base.copy()
            ed_base.update(ed)
            if 'phases' in ed:
                for phase in ed['phases']:
                    record = ed_base.copy()
                    record.update(phase)
                    records.append(record)
            else:
                records.append(ed_base)
    return pd.DataFrame(records)

# Aplatir les données
flattened_df = flatten_data(df)
# Trier les données par date
flattened_df = flattened_df.sort_values(by='date').reset_index(drop=True)
df=flattened_df
# Afficher les premières lignes du DataFrame aplati
df.head()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Function to create sequences for LSTM
def create_sequences(data, time_steps=1):
    sequences = []
    labels = []
    for i in range(len(data) - time_steps):
        sequences.append(data[i:(i + time_steps)])
        labels.append(data[i + time_steps])
    return np.array(sequences), np.array(labels)

# Prepare data for LSTM and detect anomalies
def detect_anomalies(df, channel_id):
    # Calculate total mean and std for the entire dataset upfront
    mean_total_active_power = df['active_pow'].mean()
    std_total_active_power = df['active_pow'].std() # Calculate standard deviation here

    for phase in [1, 2, 3]:
        phase_data = df[(df['channel_id'] == channel_id) & (df['phase_id'] == phase)][['date', 'active_pow']]

        # Check if enough data is available
        if len(phase_data) < 30:
            print(f"Not enough data for channel {channel_id}, phase {phase}.")
            continue

        # Normalize data
        scaler = MinMaxScaler()
        phase_data_scaled = scaler.fit_transform(phase_data[['active_pow']])

        # Split data into sequences for LSTM
        time_steps = 30
        X, y = create_sequences(phase_data_scaled, time_steps)

        # Build and train LSTM model
        model = Sequential()
        model.add(LSTM(units=64, return_sequences=True, input_shape=(time_steps, 1)))
        model.add(Dropout(0.2))
        model.add(LSTM(units=32, return_sequences=False))
        model.add(Dropout(0.2))
        model.add(Dense(units=1))

        model.compile(optimizer='adam', loss='mae')

        # Training
        history = model.fit(X, y, epochs=10, batch_size=64, validation_split=0.1, shuffle=False)

        # Predictions
        X_pred = model.predict(X)

        # Calculate reconstruction loss for each sequence
        losses = np.mean(np.abs(X_pred - y), axis=1)

        # Anomaly detection
        threshold = mean_total_active_power +   10 * std_total_active_power # Use the calculated std here
        anomalies = phase_data['active_pow'].iloc[time_steps:] > threshold

        # Add anomalies to original DataFrame
        phase_data['anomaly'] = 0
        phase_data['anomaly'].iloc[time_steps:] = anomalies.astype(int)

        # Visualize anomalies
        plt.figure(figsize=(12, 6))
        plt.plot(phase_data['date'], phase_data['active_pow'], label='Active Power')
        plt.scatter(phase_data[phase_data['anomaly'] == 1]['date'], phase_data[phase_data['anomaly'] == 1]['active_pow'], color='red', label='Anomaly')
        plt.title(f'Anomalies in Channel {channel_id} - Phase {phase}')
        plt.xlabel('Date')
        plt.ylabel('Active Power')
        plt.legend()
        plt.show()

# Detect anomalies for each phase of channels 2 and 5
detect_anomalies(df, 5)
detect_anomalies(df, 2)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Fonction pour créer des séquences pour LSTM
def create_sequences(data, time_steps=1):
    sequences = []
    labels = []
    for i in range(len(data) - time_steps):
        sequences.append(data[i:(i + time_steps)])
        labels.append(data[i + time_steps])
    return np.array(sequences), np.array(labels)

# Filtrer les données pour channel_id 4
channel_id_4_data = df[df['channel_id'] == 4]

# Conversion de la colonne 'date' en type datetime si ce n'est pas déjà fait
channel_id_4_data['date'] = pd.to_datetime(channel_id_4_data['date'])

# Extraire le jour de la semaine
channel_id_4_data['day_of_week'] = channel_id_4_data['date'].dt.dayofweek

# Définir les noms des jours de la semaine
days = ['Day0', 'Day1', 'Day2', 'Day3', 'Day4', 'Day5', 'Day6']

# Paramètres du LSTM
time_steps = 30
epochs = 10
batch_size = 64

# Créer une visualisation pour chaque jour de la semaine
for i in range(7):
    # Filtrer les données pour le jour de la semaine spécifique
    daily_data = channel_id_4_data[channel_id_4_data['day_of_week'] == i]

    # Calculer la moyenne et l'écart type de la puissance active totale pour ce jour
    mean_power = daily_data['tot_active_pow'].mean()
    std_power = daily_data['tot_active_pow'].std()

    # Normalisation des données
    scaler = MinMaxScaler()
    daily_data_scaled = scaler.fit_transform(daily_data['tot_active_pow'].values.reshape(-1, 1))

    # Création des séquences pour le modèle LSTM
    X, y = create_sequences(daily_data_scaled, time_steps)

    # Création et entraînement du modèle LSTM
    model = Sequential()
    model.add(LSTM(units=64, return_sequences=True, input_shape=(time_steps, 1)))
    model.add(Dropout(0.2))
    model.add(LSTM(units=32, return_sequences=False))
    model.add(Dropout(0.2))
    model.add(Dense(units=1))

    model.compile(optimizer='adam', loss='mae')
    history = model.fit(X, y, epochs=epochs, batch_size=batch_size, validation_split=0.1, shuffle=False)

    # Prédictions et calcul des erreurs de reconstruction
    X_pred = model.predict(X)
    losses = np.mean(np.abs(X_pred - y), axis=1)

    # Détection des anomalies basée sur un seuil dynamique
    # On utilise par exemple 5 fois l'écart type au-dessus de la moyenne
    threshold = mean_power + 5 * std_power
    anomalies = daily_data['tot_active_pow'].values[time_steps:] > threshold

    # Visualisation des anomalies détectées
    plt.figure(figsize=(12, 6))
    plt.plot(daily_data['date'], daily_data['tot_active_pow'], label='Total Active Power')
    plt.axhline(mean_power, color='orange', linestyle='--', label='Mean Power')
    plt.axhline(threshold, color='green', linestyle='--', label='Anomaly Threshold')
    plt.scatter(daily_data['date'].iloc[time_steps:][anomalies],
                daily_data['tot_active_pow'].iloc[time_steps:][anomalies],
                color='red', label='Anomalies')
    plt.title(f'Anomalies in Total Active Power for {days[i]} (Channel ID 4)')
    plt.xlabel('Date')
    plt.ylabel('Total Active Power')
    plt.legend()
    plt.xticks(rotation=45)
    plt.show()
```
